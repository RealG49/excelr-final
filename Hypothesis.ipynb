```python
import numpy as np
import scipy.stats as stats
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Given data
sample_mean = 48500
pop_mean = 50000
std_dev = 4500
n = 50

# Z-test calculation
z_statistic = (sample_mean - pop_mean) / (std_dev / np.sqrt(n))
p_value = stats.norm.sf(abs(z_statistic)) * 2  # Two-tailed test

print(f"Z-Statistic: {z_statistic:.3f}")
print(f"P-Value: {p_value:.5f}")

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("Reject H0: The average salary is significantly different from ₹50,000.")
else:
    print("Fail to Reject H0: No significant difference in salaries.")
```

```python
# Given data
diet_A = [5.2, 5.5, 4.8, 5.0, 5.4, 5.6, 4.9, 5.3, 5.1, 5.7]
diet_B = [4.6, 4.9, 4.3, 4.4, 4.8, 4.2, 4.5, 4.7, 4.1, 4.9]

# Independent T-test
t_stat, p_value = stats.ttest_ind(diet_A, diet_B)

print(f"T-Statistic: {t_stat:.3f}")
print(f"P-Value: {p_value:.5f}")

# Interpretation
if p_value < alpha:
    print("Reject H0: There is a significant difference in weight loss between Diet A and Diet B.")
else:
    print("Fail to Reject H0: No significant difference in weight loss.")
```

```python
# Contingency table (Product preference by gender)
data = np.array([[40, 60],  # Product A
                 [50, 30],  # Product B
                 [30, 40]]) # Product C

chi2_stat, p_value, dof, expected = stats.chi2_contingency(data)

print(f"Chi-Square Statistic: {chi2_stat:.3f}")
print(f"P-Value: {p_value:.5f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies:")
print(expected)

# Interpretation
if p_value < alpha:
    print("Reject H0: There is a significant association between gender and product preference.")
else:
    print("Fail to Reject H0: No significant association between gender and product preference.")
```

```python
# Given data
method_1 = [75, 80, 85, 78, 82]
method_2 = [72, 76, 79, 74, 70]
method_3 = [85, 88, 92, 90, 87]

# ANOVA test
f_stat, p_value = stats.f_oneway(method_1, method_2, method_3)

print(f"F-Statistic: {f_stat:.3f}")
print(f"P-Value: {p_value:.5f}")

# Interpretation
if p_value < alpha:
    print("Reject H0: At least one teaching method has a significantly different effect on scores.")
else:
    print("Fail to Reject H0: No significant difference between teaching methods.")
```

```python
plt.figure(figsize=(10, 5))
sns.histplot(diet_A, color="blue", label="Diet A", kde=True)
sns.histplot(diet_B, color="red", label="Diet B", kde=True)
plt.legend()
plt.title("Distribution of Weight Loss for Diet A vs. Diet B")
plt.show()
```

```python
data = pd.DataFrame({
    "Method 1": method_1,
    "Method 2": method_2,
    "Method 3": method_3
})

plt.figure(figsize=(8, 6))
sns.boxplot(data=data)
plt.title("Boxplot of Exam Scores for Different Teaching Methods")
plt.show()
```
